unpivot을 하는 이뉴는 새로운(다른방향) pivot을 하기 위해 쓰는 것이다.

```python
import pandas as pd

df1 = pd.read_excel("D00EXAMPLE.xlsx", sheet_name=0).set_index("노래")
df1
```

```python
udf1 = df1.stack().reset_index()
udf1.columns = ["심사자", "참가자", "점수"]
udf1
```

```python
df2 = pd.read_excel("D00EXAMPLE.xlsx", sheet_name=3)
df2
```

```python
pdf1 = udf1.merge(df2, how="left")
pdf1
```

```python
pdf1 = pdf1.pivot_table("점수", columns="성별", index="참가자", aggfunc="mean")
pdf1.to_clipboard()
```

```python
# 전체코드
import pandas as pd

df1 = pd.read_excel("D00EXAMPLE.xlsx").set_index("노래")
df2 = pd.read_excel("D00EXAMPLE.xlsx", sheet_name=3)
udf1 = df1.stack().reset_index()
udf1.columns = ["심사자", "참가자", "점수"]
pdf1 = udf1.merge(df2, how="left")
pdf1 = pdf1.pivot_table("점수", index="참가자", columns="성별", aggfunc="mean")
pdf1.to_clipboard()
```
