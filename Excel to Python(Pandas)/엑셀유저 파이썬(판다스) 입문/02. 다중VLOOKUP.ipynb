# 1. 엑셀 읽어오기 : read_excel()

```python
import pandas as pd

df1 = pd.read_excel("E02EXAMPLE.xlsx", sheet_name=1 )
df1
```

```python
df2 = pd.read_excel("E02EXAMPLE.xlsx", sheet_name=2)
df2
```

# 2. 데이터프레임 병합 : merge()

```python
df3 = df1.merge(df2, how="left")
df3
```

# 3. 클립보드에 복사 : to_clipboard()

```python
df3.to_clipboard(index=False)
```

# 4. 전체코드

```python
import pandas as pd
df1 = pd.read_excel("E02EXAMPLE.xlsx", sheet_name=1)
df2 = pd.read_excel("E02EXAMPLE.xlsx", sheet_name=2)
df3 = df1.merge(df2, how='left')
df3.to_clipboard(index=False)
```
