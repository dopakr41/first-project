# 1. 엑셀 읽어오기 : read_excel()

```python
import pandas as pd

df1 = pd.read_excel("E04EXAMPLE.xlsx", sheet_name=1)
df1
```

# 2. 결측치 존재확인 : is_null() or info()

```python
df1.isnull()
```

```python
df1.isnull().sum() # 결측치가 True(1)로 sum으로 집계하여 NaN(결측치)이 몇개인지 알 수 있음
```

```python
df1.isnull().sum().sum() > 0 # 0보다 크면 결측치가 있는지 판단할 수 있음
```

```python
df1.info()
```

# 3. 결측치를 특정값으로처리 : fillna()

```python
df2 = df1.fillna(0) # 결측치를 0 으로 바꿈 (디폴트 값 : 0)
df2
```

# 4. 결측치가 있는 행 삭제 : dropna()

```python
df3 = df1.dropna() # index(행)의 결측치를 삭제
df3
```

```python
df1.dropna(axis=1) # columns(열)의 결측치를 삭제
```

```python
df1
```

# 5. 클립보드에 복사 : to_clipboard()

```python
df2.to_clipboard(index=False)
```

```python
df3.to_clipboard(index=False)
```

# 6. 전체코드

```python
import pandas as pd

df1 = pd.read_excel("E04EXAMPLE.xlsx", sheet_name=1)
df2 = df1.fillna(0)
df3 = df1.dropna()
df2.to_clipboard(index=False)
df3.to_clipboard(index=False)
```
