BOOLEAN masking   
논리연산자 : &(and), |(or), ~(not)  
관계연산자 : ==, !=, >, <, >=, <=  
함수 : df.mask(cond, 참일때의 값)

# 1. 엑셀 읽어오기 : read_excel()

```python
import pandas as pd

df1 = pd.read_excel("E06EXAMPLE.xlsx", sheet_name=1)
df1
```

# 2. 엑셀의 if기능 : mask

```python
cond1 = df1["점수"].isnull()
cond1
```

```python
df1["참석여부"] = df1["점수"].mask(cond1, "X").mask(~cond1, "O")
df1
```

# 3. 엑셀의 다중if : 다중mask

```python
cond2 = df1["점수"] > 70
cond3 = df1["점수"] > 90
df1["성적"] = df1["점수"].mask(~cond2, "C").mask(cond2, "B").mask(cond3, "A")  # mask(cond2 & ~cond3, "B")
df1
```

# 4. 클립보드에 복사 : to_clipboard

```python
df1.to_clipboard(index=False)
```

# 5. 전체코드

```python
import pandas as pd 

df1 = pd.read_excel("E06EXAMPLE.xlsx", sheet_name=1)
cond1 = df1["점수"].isnull()
df1["참석여부"] = df1["점수"].mask(cond1, "X").mask(~cond1, "O")
cond2 = df1["점수"] > 70
cond3 = df1["점수"] > 90
df1["성적"] = df1["점수"].mask(~cond2, "C").mask(cond2, "B").mask(cond3, "A")
df1.to_clipboard(index=False)
```
