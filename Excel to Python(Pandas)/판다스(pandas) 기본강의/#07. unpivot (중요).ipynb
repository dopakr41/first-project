```python
import pandas as pd
val = [[83, 93, 99], [77, 88, 91], [92, 78, 95]]
idx = ["권보아", "김범수", "박효신"]
df = pd.DataFrame(val, index=idx, columns=["국어", "수학", "과학"])
val1 = [[20, 10], [5, 25], [22, 8], [12, 18]]
idx1 = [["1반", "1반", "2반", "2반"], ["남", "여","남","여"]]
df1 = pd.DataFrame(val1,index=idx1, columns=["도시락", "샌드위치"])
df2 = df1.copy().unstack()
df3 = df1.copy()
df3["업체"] = ["CJ푸드", "신세계", "신세계", "CJ푸드"]
df4 = df3.reset_index().rename(columns={"level_0":"반", "level_1":"성별"}).copy()
```

판다스는 stack과 melt로 언피벗을 한다.

1.  melt : DataFrame의 일부 columns은 unpivot하지 않는 경우 씀
2.  stack : DataFrame의 columns를 index로 바꿈 => stack + reset_index로 unpivot을 할 수 있음  
df.stack(level=숫자, dropna=bool)


```python
df = df.stack().reset_index() # reset_index로 columns명이 무작위로 생성 됨.
df.columns = ["이름", "과목", "점수"] # columns 명을 지정 함.
df.head(4)
```

```python
df1 # index가 멀티인덱스인 DataFrame
```

```python
df1 = df1.stack().reset_index()
df1.columns = ["반", "성별", "제품", "수량"]
df1
```

```python
df2 # columns가 멀티인덱스인 DataFrame
```

```python
# df2.stack().stack() # staxk을 두번 씀 
df2 = df2.stack(level=[0,1]).reset_index()
df2.columns = ["반", "제품", "성별", "수량"]
df2
```

```python
df3 # 도시락, 샌드위치만 unpivot이 되어야 함 => 업체를 index로 지정하고 stack을 해주면 됨
# df.set_index(값, dropna=bool, append=bool, inplace=bool)
```

```python
df3 = df3.set_index("업체", append=True).stack().reset_index()
df3.columns = ["반", "성별", "업체", "제품", "수량"]
df3
```
