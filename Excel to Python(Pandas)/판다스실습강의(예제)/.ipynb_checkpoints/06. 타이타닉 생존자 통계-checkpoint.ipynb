```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/TeamLab/machine_learning_from_scratch_with_python/master/code/ch12/titanic/train.csv")
df
```

```python
df1 = df.iloc[:,[1,2,4,5]]
df1.head(4)
```

```python
df1.info()
```

```python
df1.describe()  # object가 아닌 숫자형 열들의 정보를 보여줌
```

```python
df1["Survived"].unique() # 특정열의 고유값을 보여줌
```

```python
df1["Pclass"].unique()
```

```python
df1 = df1.dropna() # 결측치 drop
df1.info()
```

```python
df2 = df1.pivot_table("Survived", index="Pclass", columns="Sex", aggfunc="mean")
df2
```
