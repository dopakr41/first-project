```python
import csv
f = open('gender.csv')
data = csv.reader(f)

m = []
f = []

name = input("     : ")
for row in data:
    if '' in row[0]:
        for i in range(0,101):
            m.append(-int(row[i+3]))
            f.append(int(row[-(i+1)]))
        break
f.reverse()
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rc('font', family="Malgun Gothic")
plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(101), m, label='')
plt.barh(range(101), f, label='')
plt.legend()
plt.show()
```

```python
import csv
 
f = open('gender.csv')
data = csv.reader(f)
size = []
name = input('     : ')
for row in data :
    if name in row[0] :
        m = 0
        f = 0
        for i in range(101) :
            m += int(row[i+3])    # 3 ~ 103  
            f += int(row[i+106])  # 106 ~ 206  
        break
size.append(m)                    #     
size.append(f)                    #     
print(size)
 
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.pie(size, labels=['', ''], autopct='%.1f%%', colors=color, startangle=90)
plt.title(name+'    ')
plt.legend()
plt.show()
```
